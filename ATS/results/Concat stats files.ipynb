```python
import os
import pandas as pd
```

```python
PATH = 'stats/'
#COLS = ['start','end','RMSPE_SET']
COLS = ["start","end","RMSPE_SET", "RMSPE_MSET", "RMSPE_SEQ",
        "MAPE_SET", "MAPE_MSET", "MAPE_SEQ"]

def set_id_cols_cop(df, filename):
    aux = filename.replace('Apr21_','').replace('_h1_STATS.csv','')
    df['id'] = aux
    aux = aux.split('_')
    df['run'] = aux[0].replace('run','')
    df['constraints_type'] = aux[1]
    df['unique%'] = aux[2].replace('unique','')
    df['rep%'] = aux[3].replace('rep','').replace('-group','')
    df['group'] = aux[4]
    return df[['id','constraints_type','unique%','rep%','run','group']+COLS]

def set_id_cols(df, filename):
    aux = filename.replace('Jan22_','').replace('_h1_STATS.csv','')
    df['id'] = aux
    aux = aux.split('_')
    df['feature_composition'] = aux[0]
    df['attribute_selection'] = aux[1]
    df['feature_counting'] = aux[2]
    df['duplications'] = aux[3].replace('duplicate','')
    df['clustering_alg'] = aux[4]
    df['k'] = aux[5].replace('k','')
    df['run'] = aux[6].replace('run','')
    df['group'] = aux[7].replace('group','')
    newcols = ['id','feature_composition','attribute_selection','feature_counting',
               'duplications','clustering_alg', 'k', 'run', 'group']
    return df[newcols+COLS]
```

```python
all_data = pd.DataFrame()
files = os.listdir(PATH)
files = [f for f in files if '.csv' in f and 'Jan22' in f]

for filename in files:
    data = pd.read_csv(PATH+filename, sep=';', usecols = COLS)
    data = set_id_cols(data, filename)
    all_data = pd.concat([all_data, data])
```
